第3問 問3

    必要な知識: 自作関数、if分岐

```python
def MAISU(kingaku): #「関数の引数として与えられる金額の値をプログラムの変数kingakuに設定」
  Kouka = [1, 5, 10, 50, 100]
  maisu = 0
  nokori = kingaku
  for i in range(4, -1, -1):
    maisu = maisu + nokori // Kouka[i]
    nokori = nokori % Kouka[i]
  return maisu #「(7)行目の代わりに変数maisuの値を関数の戻り値とする」

kakaku = 46
min_maisu = 100
for tsuri in range(0, 100):
  shiharai = kakaku + tsuri
  maisu = MAISU(shiharai) + MAISU(tsuri)
  if maisu < min_maisu:
    min_maisu = maisu
print(min_maisu)
```

おまけ
*   使いやすいように質問形式にした。
*   お札にも対応した。


```python
def MAISU(kingaku):
  Kouka = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]
  maisu = 0
  nokori = kingaku
  for i in range(8, -1, -1):
    maisu = maisu + nokori // Kouka[i]
    nokori = nokori % Kouka[i]
  return maisu

kakaku = int(input('請求額を入力>> '))
min_maisu = 65535
min_s_maisu = 0
min_t_maisu = 0
min_shiharai = 0
min_tsuri = 0
for tsuri in range(0, 10000):
  shiharai = kakaku + tsuri
  maisu = MAISU(shiharai) + MAISU(tsuri)
  if maisu < min_maisu:
    min_maisu = maisu
    min_s_maisu = MAISU(shiharai)
    min_t_maisu = MAISU(tsuri)
    min_shiharai = shiharai
    min_tsuri = tsuri
print('支払額は ' + str(min_shiharai) + ' 円 (' + str(min_s_maisu) + '枚)')
print('お釣りは ' + str(min_tsuri) + ' 円 (' + str(min_t_maisu) + '枚)')
print('総やり取り枚数は ' + str(min_maisu) + ' 枚')
```
